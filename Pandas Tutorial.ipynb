#### For loop to categorize cars prices and save in the new column

```python
for i in range(0, len(cars_data2["Price"]), 1):
    if(cars_data2["Price"][i]<=8450):
        cars_data2["Price_class"][i]="Low"
    elif((cars_data2["Price"][i]>11950)):
        cars_data2["Price_class"][i]="High"
    else:
        cars_data2["Price_class"][i]="Medium"
```

```python
print(cars_data2["Price_class"])
```

#### While loop to categorize cars prices and save in the new column

```python
i=0

while i<len(cars_data2["Price"]):
    if(cars_data2["Price"][i]<=8450):
        cars_data2["Price_class"][i]="Low"
    elif((cars_data2["Price"][i]>11950)):
        cars_data2["Price_class"][i]="High"
    else:
        cars_data2["Price_class"][i]="Medium"
    i=i+1
```

## Functions

#### Single Input Single Output

```python
cars_data2.insert(11, "Age_converted", 0)
```

#### Function to convert cars age from Months to years

```python
def c_convert(val):
    val_converted = val/12
    return val_converted
```

```python
cars_data2["age_converted"]=c_convert(cars_data2['Age'])
```

```python
cars_data2["age_converted"]=round(c_convert(cars_data2['Age']), 1)
```

#### Multi Input Multi Output

```python
cars_data2.insert(12, "KM_per_month", 0)
```

#### Function to convert cars age from Months to years and to find KM per month run by car

```python
def c_convert(val1, val2):
    val_converted = val1/12
    ratio = val2/val1   
    return [val_converted, ratio]
```

```python
cars_data2["age_converted"],cars_data2["KM_per_month"] = c_convert(cars_data2['Age'], cars_data2['KM'])
```

## Aggregate Statistics (Groupby function)

```python
cars_data3 = pd.read_csv('filtered.csv')

cars_data3['count'] = 1

cars_data3.groupby(['FuelType', 'HP']).count()['count']
```

## Exploratory Data Analysis 
#### Cross Tabulation 
###### Checking the relationship between two Categorical Variables

```python
pd.crosstab(index=cars_data3['FuelType'], columns='count', dropna=True)
```

```python
pd.crosstab(index = cars_data3['Automatic'], 
            columns=cars_data3['FuelType'],
            dropna=True)
```

```python
pd.crosstab(index = cars_data3['Automatic'], 
            columns=cars_data3['FuelType'],
            normalize=True,
            dropna=True)
```

```python
pd.crosstab(index = cars_data3['Automatic'], 
            columns=cars_data3['FuelType'],
            margins=True,
            normalize=True,
            dropna=True)
```

## Conditional Probability (Row sum)

```python
pd.crosstab(index = cars_data3['Automatic'], 
            columns=cars_data3['FuelType'],
            margins=True,
            normalize='index',
            dropna=True)
```

## Conditional Probability (Column Sum)

```python
pd.crosstab(index = cars_data3['Automatic'], 
            columns=cars_data3['FuelType'],
            margins=True,
            normalize='columns',
            dropna=True)
```
